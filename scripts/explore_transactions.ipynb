# Messy Column Exploration

```python
# Load the data once for the later cells
import pandas as pd
from pathlib import Path

csv_path = Path('../data/raw/consolidated_transactions_preview.csv')
df = pd.read_csv(csv_path)
print('Loaded rows:', len(df))
print('Columns:', list(df.columns))
```

```python
# Inspect transaction IDs
print('Transaction ID sample values:')
print(df['Transaction ID'].head(20).to_list())
print('\nTransaction ID dtype:', df['Transaction ID'].dtype)
print('Transaction ID nulls:', df['Transaction ID'].isna().sum())
print('Transaction ID numeric-like values:', df['Transaction ID'].astype(str).str.match(r'^-?\d+(\.0+)?$').sum())
print('Transaction ID non-numeric values:', df['Transaction ID'].astype(str).str.match(r'^-?\d+(\.0+)?$').eq(False).sum())
```

```python
# Inspect cost columns
cost_cols = ['unit_cost', 'standard_unit_cost', 'landed_unit_cost']
for col in cost_cols:
    print(f'\n{col}:')
    print('nulls:', df[col].isna().sum())
    print('non-null sample:', df[col].dropna().head(10).tolist())
    print('min/max:', df[col].dropna().min(), df[col].dropna().max())

df[df["unit_cost"].isna()]
```

```python
# Inspect missing product and store keys
for col in ['Product index', 'Store ID']:
    print(f'\n{col}:')
    print('nulls:', df[col].isna().sum())
    print('sample null rows:')
    print(df[df[col].isna()].head(5).to_string(index=False))
```

```python
# Inspect duplicate transaction IDs
print('Duplicate transaction IDs count:', df['Transaction ID'].duplicated().sum())
print('Top duplicate transaction IDs:')
print(df['Transaction ID'].value_counts().head(15).to_string())
```

```python
# Compare standard vs landed costs
mask = df['standard_unit_cost'].notna() & df['landed_unit_cost'].notna()
print('Rows with both standard and landed costs present:', mask.sum())
print('Rows where standard != landed:', ((df.loc[mask, 'standard_unit_cost'] != df.loc[mask, 'landed_unit_cost']).sum()))
comparison = df.loc[mask, ['standard_unit_cost', 'landed_unit_cost', 'pack_size', 'unit_price']].copy()
comparison['diff'] = comparison['landed_unit_cost'] - comparison['standard_unit_cost']
print('Top differences:')
print(comparison.sort_values('diff', ascending=False).head(15).to_string(index=False))
```

```python
# Check rows where landed cost exists but standard cost is missing
print('Rows with landed cost but no standard cost:')
print(df.loc[df['landed_unit_cost'].notna() & df['standard_unit_cost'].isna(), ['unit_cost','standard_unit_cost','landed_unit_cost','pack_size','Product index','Store ID']].head(20).to_string(index=False))
```
